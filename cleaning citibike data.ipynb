```python
# Import data
jun = pd.read_csv('data/JC-202306-citibike-tripdata.csv',)
jul = pd.read_csv('data/JC-202307-citibike-tripdata.csv',)
aug = pd.read_csv('data/JC-202308-citibike-tripdata.csv',)
sep = pd.read_csv('data/JC-202309-citibike-tripdata.csv')



display(jun.head(2))
display(jul.head(2))
display(aug.head(2))
display(sep.head(2))
```

```python
# Check if all of the column titles in all of the DataFrames are exactly the same both in spelling and in order
result = (
    (list(jun.columns) == list(jul.columns)) &
    (list(jun.columns) == list(aug.columns)) &
    (list(jun.columns) == list(sep.columns))
)
print(result)
        
# Check if all of the columns have the same data types in all of the DataFrames
check=(
    (jun.dtypes == jul.dtypes) & 
    (jun.dtypes == aug.dtypes) & 
    (jun.dtypes == sep.dtypes))
print(check)
```

```python
# Check which rows have missing values

print(f"jun counts pre-.dropna():\n{jun.count()}\n")
print(f"jul counts pre-.dropna():\n{jul.count()}\n")
print(f"aug counts pre-.dropna():\n{aug.count()}\n")
print(f"sep counts pre-.dropna():{sep.count()}")
```

```python
# Remove missing values

jun.dropna(inplace=True)
jul.dropna(inplace=True)
aug.dropna(inplace=True)
sep.dropna(inplace=True)
```

```python
#check if the final df has any null values

print(jun.isnull().values.any())
print(jul.isnull().values.any())
print(aug.isnull().values.any())
print(sep.isnull().values.any())
```

```python
# Combine data
all_data = pd.concat([jun,jul, aug, sep], ignore_index=True)
display(all_data.head())
```

```python
# Check that all of the columns have the proper data types
all_data.info()
```

```python
# Convert the 'start_time' and 'stop_time' columns to the datetime data type
all_data['started_at'] = pd.to_datetime(all_data['started_at'])
all_data['ended_at'] = pd.to_datetime(all_data['ended_at'])

# Confirm the data types were changed
all_data.dtypes
```

```python
# Renaming the columns for cleaner look
all_data.rename(columns={
    'ride_id':'Ride_ID',
    'rideable_type': 'Bike_Type',
    'started_at':'Start_Time',
    'ended_at':'End_Time',
    'start_station_name':'Start_Station_Name',
    'start_station_id':'Start_Station_ID',
    'end_station_name':'End_Station_Name',
    'end_station_id':'End_Station_ID',
    'start_lat':'Start_Lattitude ',
    'start_lng':'Start_Longitude',
    'end_lat':'End_Latitude',
    'end_lng':'End_Longitude',
    'member_casual':'Member_Type'   
}, inplace=True)
```
