```python
import pandas as pd
import numpy as np
import os
```

```python
# Get the data file names
os.listdir('data')
```

```python
# Import data
jan = pd.read_csv('data/JC-202301-citibike-tripdata.csv',)
feb = pd.read_csv('data/JC-202302-citibike-tripdata.csv',)
mar = pd.read_csv('data/JC-202303-citibike-tripdata.csv',)
apr = pd.read_csv('data/JC-202304-citibike-tripdata.csv',)
may = pd.read_csv('data/JC-202305-citibike-tripdata.csv',)
jun = pd.read_csv('data/JC-202306-citibike-tripdata.csv',)
jul = pd.read_csv('data/JC-202307-citibike-tripdata.csv',)
aug = pd.read_csv('data/JC-202308-citibike-tripdata.csv',)
sep = pd.read_csv('data/JC-202309-citibike-tripdata.csv')


display(jan.head(2))
display(feb.head(2))
display(mar.head(2))
display(apr.head(2))
display(may.head(2))
display(jun.head(2))
display(jul.head(2))
display(aug.head(2))
display(sep.head(2))
```

```python
# Check if all of the column titles in all of the DataFrames are exactly the same both in spelling and in order
result = (
    (list(jan.columns) == list(feb.columns)) &
    (list(jan.columns) == list(mar.columns)) &
    (list(jan.columns) == list(apr.columns)) &
    (list(jan.columns) == list(may.columns)) &
    (list(jan.columns) == list(jun.columns)) &
    (list(jan.columns) == list(jul.columns)) &
    (list(jan.columns) == list(aug.columns)) &
    (list(jan.columns) == list(sep.columns))
)
print(result)
        
# Check if all of the columns have the same data types in all of the DataFrames
check=(
    (jan.dtypes == feb.dtypes) & 
    (jan.dtypes == mar.dtypes) & 
    (jan.dtypes == apr.dtypes) &
    (jan.dtypes == may.dtypes) & 
    (jan.dtypes == jun.dtypes) & 
    (jan.dtypes == jul.dtypes) & 
    (jan.dtypes == aug.dtypes) & 
    (jan.dtypes == sep.dtypes))
print(check)
```

```python
# Check which rows have missing values
print(f"\nmay counts pre-.dropna():\n{jan.count()}\n")
print(f"\nmay counts pre-.dropna():\n{feb.count()}\n")
print(f"\nmay counts pre-.dropna():\n{mar.count()}\n")
print(f"\nmay counts pre-.dropna():\n{apr.count()}\n")
print(f"\nmay counts pre-.dropna():\n{may.count()}\n")
print(f"jun counts pre-.dropna():\n{jun.count()}\n")
print(f"jul counts pre-.dropna():\n{jul.count()}\n")
print(f"aug counts pre-.dropna():\n{aug.count()}\n")
print(f"sep counts pre-.dropna():{sep.count()}")
```

```python
# Remove missing values
jan.dropna(inplace=True)
feb.dropna(inplace=True)
mar.dropna(inplace=True)
apr.dropna(inplace=True)
may.dropna(inplace=True)
jun.dropna(inplace=True)
jul.dropna(inplace=True)
aug.dropna(inplace=True)
sep.dropna(inplace=True)
```

```python
#check if the final df has any null values

print(jan.isnull().values.any())
print(feb.isnull().values.any())
print(mar.isnull().values.any())
print(apr.isnull().values.any())
print(may.isnull().values.any())
print(jun.isnull().values.any())
print(jul.isnull().values.any())
print(aug.isnull().values.any())
print(sep.isnull().values.any())
```

```python
# Combine data
all_data = pd.concat([jan,feb,mar,apr,may, jun,jul, aug, sep], ignore_index=True)
display(all_data.head())
```

```python
# Check that all of the columns have the proper data types
all_data.info()
```

```python
# Convert the 'start_time' and 'stop_time' columns to the datetime data type
all_data['started_at'] = pd.to_datetime(all_data['started_at'])
all_data['ended_at'] = pd.to_datetime(all_data['ended_at'])

# Confirm the data types were changed
all_data.dtypes
```

```python
# Renaming the columns for cleaner look
all_data.rename(columns={
    'ride_id':'Ride_ID',
    'rideable_type': 'Bike_Type',
    'started_at':'Start_Time',
    'ended_at':'End_Time',
    'start_station_name':'Start_Station_Name',
    'start_station_id':'Start_Station_ID',
    'end_station_name':'End_Station_Name',
    'end_station_id':'End_Station_ID',
    'start_lat':'Start_Lat ',
    'start_lng':'Start_Lng',
    'end_lat':'End_Lat',
    'end_lng':'End_Lng ',
    'member_casual':'Member_Type'   
}, inplace=True)
```

```python
all_data.to_csv('data/all_data.csv', index=False)
```
